# Exploring the sci-kit learn iris dataset

```
# import modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

# import KNN classifier
from sklearn.neighbors import KNeighborsClassifier

# import iris dataset
from sklearn.datasets import load_iris

# load iris dataset
iris = load_iris()

# load dataset into pandas dataframe
feature_data = iris.data
df = pd.DataFrame(feature_data, columns=iris.feature_names)

# set target
target_data = iris.target
df['target'] = target_data
```

```
# check if there are empty records
df.isnull().sum()
```

```
# describe dataframe
df.describe()
```

```
# get number of samples in each group
df.groupby('target').size()
```

```
# create x and y variables to use in classification model
x = feature_data
y = target_data

# create KNN classifier model
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x,y)
```

```
df.tail()
```

```
# test prediction on row 149
knn.predict([[5.9, 3, 5.1, 1.8]])
```

```
df.head()
```

```
# test prediction on row 4
knn.predict([[5, 3.6, 1.4, 0.2]])
```

```
# create scatterplot to visualise distribution
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='target')
plt.legend(title='variety', loc='upper left', labels=iris.target_names)
```
